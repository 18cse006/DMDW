```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
path=('https://raw.githubusercontent.com/18cse006/DMDW/main/Car_sales.csv')
```

```
data=pd.read_csv(path)
```

```
import seaborn as sns
```

```
data.head()
```

```
data.dropna(axis=0,inplace=True)
data.isnull().sum() 
```

Data Visuliazation

```
plt.figure(figsize=(12,5))
plt.scatter(data['Model'],data['Horsepower'],c='green')
plt.xlabel('model')
plt.ylabel('Horsepower')
plt.show()
```

```
plt.hist(data['Manufacturer'],edgecolor='white',bins=5)
plt.title('Histogram of Manufacturer')
plt.xlabel('model')
plt.ylabel('Horsepower')
plt.show()
```

```
sns.set(style='darkgrid')
sns.regplot(x=data['Price_in_thousands'],y=data['Fuel_efficiency'])
```

```
sns.set(style='darkgrid')
sns.regplot(x=data['Price_in_thousands'],y=data['Fuel_efficiency'],marker='*',fit_reg=False)
```

```
sns.distplot(data['Price_in_thousands'])
```

```
sns.distplot(data['Price_in_thousands'],kde=False) #kernal desity estimater(kde)
```

```
sns.distplot(data['Price_in_thousands'],kde=False,bins=5)
```

```
plt.figure(figsize=(12,5))
sns.countplot(x=data['Manufacturer'],data=data)
```

Box plot

```
sns.boxplot(y='Price_in_thousands',data=data)
```
